```python
import gtts, re, os, time
```

```python
f = open("result-finale.txt",'r')
inbox = f.readlines()
```

```python
outdir = '.\sk_audio'
```

```python
inbox
```

```python
def process(ffn):
    ffn.replace("\n","")
    ffn = re.sub(r"[^a-zA-Z]","",ffn)
    try:
        af = gtts.gTTS(ffn,lang='id')
        af.save(os.path.join(outdir,ffn+".mp3"))
    except:
        print("ERROR - ", ffn)
    print(ffn)
    time.sleep(2)
```

```python
len(inbox)
```

```python
for idx,i in enumerate(inbox):
    print(idx)
    process(i)
```
